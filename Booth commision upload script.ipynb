```python
#date_range (date_list)
def date_range():
    from_date = input("from_date_format(yyyy-mm-dd :)")
    to_date = input("to_date_formate(yyyy-mm-dd :)")
    date_list = pd.date_range(start=from_date,end=to_date)
    return date_list

# tm_std_fcm_complete_report
def tm_std_fcm_report():
    date_list = ['2020-5-04']
    data_dict = {
        "product":{
            "tm":{
                "resived_from_diary": 0,
                "leakage" : 0,
                "returned" : 0,
                "net_recived" : 0
            },
            "std":{
                "resived_from_diary": 0,
                "net_litre" : 0,
                "leakage" : 0,
                "returned" : 0
            },
            "fcm":{
                "resived_from_diary": 0,
                "net_litre" : 0,
                "leakage" : 0,
                "returned" : 0
            }
        },
        "union_sale" : {},
        "coimbatore_account" : {
            "Agent + Pvt Parlour (cash)":{
                "tm" : 0,
                "std" : 0,
                "fcm" : 0
            },
            "Agent + Pvt Parlour (card)" : {
                "tm" : 0,
                "std" : 0,
                "fcm" : 0
            },
            "Ownparlour (cash)" : {
                "tm" : 0,
                "std" : 0,
                "fcm" : 0
            },
            "Ownparlour (card)" : {
                "tm" : 0,
                "std" : 0,
                "fcm" : 0
            },
            "Private ins" : {
                "tm" : 0,
                "std" : 0,
                "fcm" : 0
            },
            "Govt. ins" : {
                "tm" : 0,
                "std" : 0,
                "fcm" : 0
            }
        }
    }
    tm_std_fcm_diary_sale_obj = DailySessionllyBusinessllySale.objects.filter(delivery_date__in=date_list)
    tm_std_fcm_sale_obj = DailySessionllyBusinessllySale.objects.filter(delivery_date__in=date_list,sold_to__in = ["Agent","ICustomer"])
    tm_std_fcm_leake_obj = DailySessionllyBusinessllySale.objects.filter(delivery_date__in=date_list,sold_to="Leakage")
   
    #dairy sale (icus + agent + leakage)
    tm_sale = tm_std_fcm_diary_sale_obj.aggregate(Sum('tm500_litre'))['tm500_litre__sum']
    std_sale = tm_std_fcm_diary_sale_obj.aggregate(Sum('std250_litre'))["std250_litre__sum"] + tm_std_fcm_diary_sale_obj.aggregate(Sum('std500_litre'))["std500_litre__sum"]
    fcm_sale = tm_std_fcm_diary_sale_obj.aggregate(Sum('fcm500_litre'))["fcm500_litre__sum"] + tm_std_fcm_diary_sale_obj.aggregate(Sum('fcm1000_litre'))["fcm1000_litre__sum"]

    data_dict["product"]["tm"]["resived_from_diary"] = tm_sale
    data_dict["product"]["std"]["resived_from_diary"] = std_sale
    data_dict["product"]["fcm"]["resived_from_diary"] = fcm_sale

    #net_recived (icus + agent)
    tm_sale = tm_std_fcm_sale_obj.aggregate(Sum('tm500_litre'))['tm500_litre__sum']
    std_sale = tm_std_fcm_sale_obj.aggregate(Sum('std250_litre'))["std250_litre__sum"] + tm_std_fcm_sale_obj.aggregate(Sum('std500_litre'))["std500_litre__sum"]
    fcm_sale = tm_std_fcm_sale_obj.aggregate(Sum('fcm500_litre'))["fcm500_litre__sum"] + tm_std_fcm_sale_obj.aggregate(Sum('fcm1000_litre'))["fcm1000_litre__sum"]

    data_dict["product"]["tm"]["net_recived"] = tm_sale
    data_dict["product"]["std"]["net_recived"] = std_sale
    data_dict["product"]["fcm"]["net_recived"] = fcm_sale

    #leakage
    lk_tm_sale = tm_std_fcm_leake_obj.aggregate(Sum('tm500_litre'))['tm500_litre__sum']
    lk_std_sale = tm_std_fcm_leake_obj.aggregate(Sum('std250_litre'))["std250_litre__sum"] + tm_std_fcm_leake_obj.aggregate(Sum('std500_litre'))["std500_litre__sum"]
    lk_fcm_sale = tm_std_fcm_leake_obj.aggregate(Sum('fcm500_litre'))["fcm500_litre__sum"] + tm_std_fcm_leake_obj.aggregate(Sum('fcm1000_litre'))["fcm1000_litre__sum"]

    data_dict["product"]["tm"]["leakage"] = lk_tm_sale
    data_dict["product"]["std"]["leakage"] = lk_std_sale
    data_dict["product"]["fcm"]["leakage"] = lk_fcm_sale

    union_list = ["COIMBATORE Union","NILGIRIS Union","TIRUPPUR Union"]

    for union in union_list:
        if not union in data_dict["union_sale"]:
            data_dict["union_sale"][union] = {
                "tm" : 0,
                "std" : 0,
                "fcm" : 0
            }
        tm_std_fcm_union_obj = DailySessionllyBusinessllySale.objects.filter(delivery_date__in=date_list,union=union,sold_to__in = ["Agent","ICustomer"])

        # union sale
        union_tm_sale = tm_std_fcm_union_obj.aggregate(Sum('tm500_litre'))['tm500_litre__sum']
        union_std_sale = tm_std_fcm_union_obj.aggregate(Sum('std250_litre'))["std250_litre__sum"] + tm_std_fcm_union_obj.aggregate(Sum('std500_litre'))["std500_litre__sum"]
        union_fcm_sale = tm_std_fcm_union_obj.aggregate(Sum('fcm500_litre'))["fcm500_litre__sum"] + tm_std_fcm_union_obj.aggregate(Sum('fcm1000_litre'))["fcm1000_litre__sum"]

        data_dict["union_sale"][union]["tm"] = union_tm_sale
        data_dict["union_sale"][union]["std"] = union_std_sale
        data_dict["union_sale"][union]["fcm"] = union_fcm_sale
       
    #1.Agent + Pvt Parlour (cash)
    tm_std_fcm_cbe_obj = DailySessionllyBusinessllySale.objects.filter(delivery_date__in=date_list,union="COIMBATORE Union",sold_to = "Agent",business_type_id__in=[1,2])

    # union sale
    cbe_tm_sale = tm_std_fcm_cbe_obj.aggregate(Sum('tm500_litre'))['tm500_litre__sum']
    cbe_std_sale = tm_std_fcm_cbe_obj.aggregate(Sum('std250_litre'))["std250_litre__sum"] + tm_std_fcm_cbe_obj.aggregate(Sum('std500_litre'))["std500_litre__sum"]
    cbe_fcm_sale = tm_std_fcm_cbe_obj.aggregate(Sum('fcm500_litre'))["fcm500_litre__sum"] + tm_std_fcm_cbe_obj.aggregate(Sum('fcm1000_litre'))["fcm1000_litre__sum"]

    data_dict["coimbatore_account"]["Agent + Pvt Parlour (cash)"]["tm"] = cbe_tm_sale
    data_dict["coimbatore_account"]["Agent + Pvt Parlour (cash)"]["std"] = cbe_std_sale
    data_dict["coimbatore_account"]["Agent + Pvt Parlour (cash)"]["fcm"] = cbe_fcm_sale
   
    #2.Agent + Pvt Parlour (card)
    tm_std_fcm_cbe_obj = DailySessionllyBusinessllySale.objects.filter(delivery_date__in=date_list,union="COIMBATORE Union",sold_to = "ICustomer",business_type_id__in=[1,2])

    # union sale
    cbe_tm_sale = tm_std_fcm_cbe_obj.aggregate(Sum('tm500_litre'))['tm500_litre__sum']
    cbe_std_sale = tm_std_fcm_cbe_obj.aggregate(Sum('std250_litre'))["std250_litre__sum"] + tm_std_fcm_cbe_obj.aggregate(Sum('std500_litre'))["std500_litre__sum"]
    cbe_fcm_sale = tm_std_fcm_cbe_obj.aggregate(Sum('fcm500_litre'))["fcm500_litre__sum"] + tm_std_fcm_cbe_obj.aggregate(Sum('fcm1000_litre'))["fcm1000_litre__sum"]

    data_dict["coimbatore_account"]["Agent + Pvt Parlour (card)"]["tm"] = cbe_tm_sale
    data_dict["coimbatore_account"]["Agent + Pvt Parlour (card)"]["std"] = cbe_std_sale
    data_dict["coimbatore_account"]["Agent + Pvt Parlour (card)"]["fcm"] = cbe_fcm_sale
   
    #3.Ownparlour (cash)
    tm_std_fcm_cbe_obj = DailySessionllyBusinessllySale.objects.filter(delivery_date__in=date_list,union="COIMBATORE Union",sold_to = "Agent",business_type_id=3)

    # union sale
    cbe_tm_sale = tm_std_fcm_cbe_obj.aggregate(Sum('tm500_litre'))['tm500_litre__sum']
    cbe_std_sale = tm_std_fcm_cbe_obj.aggregate(Sum('std250_litre'))["std250_litre__sum"] + tm_std_fcm_cbe_obj.aggregate(Sum('std500_litre'))["std500_litre__sum"]
    cbe_fcm_sale = tm_std_fcm_cbe_obj.aggregate(Sum('fcm500_litre'))["fcm500_litre__sum"] + tm_std_fcm_cbe_obj.aggregate(Sum('fcm1000_litre'))["fcm1000_litre__sum"]

    data_dict["coimbatore_account"]["Ownparlour (cash)"]["tm"] = cbe_tm_sale
    data_dict["coimbatore_account"]["Ownparlour (cash)"]["std"] = cbe_std_sale
    data_dict["coimbatore_account"]["Ownparlour (cash)"]["fcm"] = cbe_fcm_sale
   
    #4.Ownparlour (card)
    tm_std_fcm_cbe_obj = DailySessionllyBusinessllySale.objects.filter(delivery_date__in=date_list,union="COIMBATORE Union",sold_to = "ICustomer",business_type_id=3)

    # union sale
    cbe_tm_sale = tm_std_fcm_cbe_obj.aggregate(Sum('tm500_litre'))['tm500_litre__sum']
    cbe_std_sale = tm_std_fcm_cbe_obj.aggregate(Sum('std250_litre'))["std250_litre__sum"] + tm_std_fcm_cbe_obj.aggregate(Sum('std500_litre'))["std500_litre__sum"]
    cbe_fcm_sale = tm_std_fcm_cbe_obj.aggregate(Sum('fcm500_litre'))["fcm500_litre__sum"] + tm_std_fcm_cbe_obj.aggregate(Sum('fcm1000_litre'))["fcm1000_litre__sum"]

    data_dict["coimbatore_account"]["Ownparlour (card)"]["tm"] = cbe_tm_sale
    data_dict["coimbatore_account"]["Ownparlour (card)"]["std"] = cbe_std_sale
    data_dict["coimbatore_account"]["Ownparlour (card)"]["fcm"] = cbe_fcm_sale
   
    #5.Private ins
    tm_std_fcm_cbe_obj = DailySessionllyBusinessllySale.objects.filter(delivery_date__in=date_list,union="COIMBATORE Union",sold_to__in = ["Agent","ICustomer"],business_type_id=4)

    # union sale
    cbe_tm_sale = tm_std_fcm_cbe_obj.aggregate(Sum('tm500_litre'))['tm500_litre__sum']
    cbe_std_sale = tm_std_fcm_cbe_obj.aggregate(Sum('std250_litre'))["std250_litre__sum"] + tm_std_fcm_cbe_obj.aggregate(Sum('std500_litre'))["std500_litre__sum"]
    cbe_fcm_sale = tm_std_fcm_cbe_obj.aggregate(Sum('fcm500_litre'))["fcm500_litre__sum"] + tm_std_fcm_cbe_obj.aggregate(Sum('fcm1000_litre'))["fcm1000_litre__sum"]

    data_dict["coimbatore_account"]["Private ins"]["tm"] = cbe_tm_sale
    data_dict["coimbatore_account"]["Private ins"]["std"] = cbe_std_sale
    data_dict["coimbatore_account"]["Private ins"]["fcm"] = cbe_fcm_sale
   
    #6.Govt. ins
    tm_std_fcm_cbe_obj = DailySessionllyBusinessllySale.objects.filter(delivery_date__in=date_list,union="COIMBATORE Union",sold_to__in = ["Agent","ICustomer"],business_type_id=5)

    # union sale
    cbe_tm_sale = tm_std_fcm_cbe_obj.aggregate(Sum('tm500_litre'))['tm500_litre__sum']
    cbe_std_sale = tm_std_fcm_cbe_obj.aggregate(Sum('std250_litre'))["std250_litre__sum"] + tm_std_fcm_cbe_obj.aggregate(Sum('std500_litre'))["std500_litre__sum"]
    cbe_fcm_sale = tm_std_fcm_cbe_obj.aggregate(Sum('fcm500_litre'))["fcm500_litre__sum"] + tm_std_fcm_cbe_obj.aggregate(Sum('fcm1000_litre'))["fcm1000_litre__sum"]

    data_dict["coimbatore_account"]["Govt. ins"]["tm"] = cbe_tm_sale
    data_dict["coimbatore_account"]["Govt. ins"]["std"] = cbe_std_sale
    data_dict["coimbatore_account"]["Govt. ins"]["fcm"] = cbe_fcm_sale

    return data_dict

#function_call
tm_std_fcm_report()
```
