```python
# import library
import pandas as pd
import numpy as np
import requests
import io

try:
    from StringIO import StringIO
except ImportError:
    
    from io import StringIO
    from io import BytesIO
from decimal import Decimal
google_sheet_url = 'https://docs.google.com/spreadsheets/d/1nDPjYd_y9IafqFe8xP9EdVUSXDTlK7bF8yBuDOfVlQ8/export?format=xlsx'
r = requests.get(google_sheet_url)
data = r.content
df = pd.read_excel(BytesIO(data),sheet_name='Sheet2')
df = df.fillna(0)
missing_list = []
available_agent_code = []
mismatch_agent_codes = []
df.head()
```

```python
for index, row, in df.iterrows():
    if BusinessAgentMap.objects.filter(business__code=row['Booth']).exists():
        if row['AgentCode'] in available_agent_code:
            mismatch_agent_codes.append(row['AgentCode'])
            continue
        if Agent.objects.filter(agent_code=row['AgentCode']).exists():
            agent_obj = Agent.objects.get(agent_code=row['AgentCode'])
            agent_obj.agent_code = str(index) + '-' + str(row['AgentCode'])
            agent_obj.save()
        business_agent_obj = BusinessAgentMap.objects.get(business__code=row['Booth'])
        if not row['Deposit'] == 0:
            business_agent_obj.deposit_amount = row['Deposit']
        business_agent_obj.save()
        
#       Update booth details
        business_obj = business_agent_obj.business
        if not row['Address'] == 0:
            business_obj.address = row['Address']
        business_obj.save()
        
#       update agent details
        agent_obj = business_agent_obj.agent
        agent_obj.agent_code = row['AgentCode']
        if not row['Aadhar'] == 0:
            agent_obj.aadhar_number = row['Aadhar']
        if not row['Address'] == 0:
            agent_obj.communication_address = row['Address']
        if not row['PAN'] == 0:
            agent_obj.pan_number = row['PAN']
        agent_obj.save()
        
#       update agent bank details
        agent_bank_obj = AgentBankDetail.objects.get(agent=agent_obj)
        if not row['BANK'] == 0:
            agent_bank_obj.bank = row['BANK']
        if not row['BRANCH'] == 0:
            agent_bank_obj.branch = row['BRANCH']
        if not row['BankAc'] == 0:
            agent_bank_obj.account_number = row['BankAc']
        if not row['IFSCCode'] == 0:
            agent_bank_obj.ifsc_code = row['IFSCCode']
        agent_bank_obj.save()
        
#       update user profile details
        user_profile_obj = agent_obj.user_profile
        if not row['Address'] == 0:
            user_profile_obj.street = row['Address']
        if not row['Mobile'] == 0:
            user_profile_obj.mobile = row['Mobile']
        if not row['MICRCode'] == 0:
            user_profile_obj.micr_code = row['MICRCode']
        user_profile_obj.save()
        
#       update user details
        user_obj = user_profile_obj.user
        name = row['Agent'].split('.')
        first_name = name[0]
        if len(name) == 2:
            last_name = name[1]
        else:
            last_name = ''
        user_obj.first_name = first_name[0:30]
        user_obj.last_name = last_name
        user_obj.username = business_obj.code
        user_obj.save()
        print('Booth code - {} Details Updated'.format(row['Booth']))
        available_agent_code.append(row['AgentCode'])
    else:
        print('Booth code - {} Not available in DB'.format(row['Booth']))
        missing_list.append(row['Booth'])
        continue
```
