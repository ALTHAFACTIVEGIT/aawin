```python
date_list = ["2020-05-20"]#,"2020-05-19","2020-05-18","2020-05-17","2020-05-16"]

data_dict = {}

for business in Business.objects.filter():
    data_dict[business.name] = {}
    for date in date_list:
        if not date in data_dict[business.name]:
            data_dict[business.name][date] = 0
        mor_sale_obj = DailySessionllyBusinessllySale.objects.filter(delivery_date=date,business_id=business.id)
        if mor_sale_obj:
            milk_total = mor_sale_obj.aggregate(Sum('tm500_litre'))['tm500_litre__sum'] + mor_sale_obj.aggregate(Sum('std250_litre'))["std250_litre__sum"] + mor_sale_obj.aggregate(Sum('std500_litre'))["std500_litre__sum"] + mor_sale_obj.aggregate(Sum('fcm500_litre'))["fcm500_litre__sum"]
            data_dict[business.name][date] = milk_total
            
df = pd.DataFrame(data_dict)
df_t = df.T
df_t
df_t.to_excel("booth_wise_milk_total_morning.xlsx")
data_dict
```
